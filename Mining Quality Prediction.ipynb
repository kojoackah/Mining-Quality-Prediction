```python
df_input = mine_df.drop(columns = '% Silica Concentrate')
df_target = mine_df['% Silica Concentrate']
```

```python
df_target.shape
```

```python
df_input = np.array(df_input)
df_target = np.array(df_target)
```

```python
df_target = df_target.reshape(-1,1)
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
from sklearn.pipeline import Pipeline, make_pipeline
```

```python
scaler_x = StandardScaler()
X = scaler_x.fit_transform(df_input)
```

```python
scaler_y = StandardScaler()
y = scaler_y.fit_transform(df_target)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
mine_df.shape
```

```python
# training and evaluate a linear regression model
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, accuracy_score
```

```python
LinearRegression_model = LinearRegression()
```

```python
LinearRegression_model.fit(X_train,y_train)
```

```python
accuracy_LinearRegression = LinearRegression_model.score(X_test,y_test)
```

```python
accuracy_LinearRegression
```

```python
# decision tree

from sklearn.tree import DecisionTreeRegressor
DecisionTree_model =DecisionTreeRegressor()
DecisionTree_model.fit(X_train, y_train)
```

```python
accuracy_DecisionTree = DecisionTree_model.score(X_test, y_test)
```

```python
accuracy_DecisionTree
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
accuracy_RandomForest = RandomForest_model.score(X_test, y_test)
accuracy_RandomForest
```

```python
y_predict = DecisionTree_model.predict(X_test)
plt.plot(y_predict,y_test, '^', color = 'r')
plt.xlabel('Predicted Values')
plt.ylabel('True Values')
```

```python
y_predict.shape
```

```python
y_predict = y_predict.reshape(-1,1)
```

```python
# going back to initial values using inverse transform

y_original_pred = scaler_y.inverse_transform(y_predict)
y_original_test = scaler_y.inverse_transform(y_test)
plt.plot(y_original_pred ,y_original_test, '^', color = 'r')
plt.xlabel('Predicted Values')
plt.ylabel('True Values')
```

```python
# This plot gives the predicted values of Silica Concentrates against the true values.
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from math import sqrt

k = X_test.shape[1]
n = len(X_test)
RMSE = float(format(np.sqrt(mean_squared_error(y_original_pred, y_original_test)),'.3f'))
MSE = mean_squared_error(y_original_pred, y_original_test)
MAE = mean_absolute_error(y_original_pred, y_original_test)
R2 = r2_score(y_original_pred, y_original_test)

adjusted_R2 = 1-(1-R2)*(n-1)/(n-k-1)

print('RMSE :', RMSE, '\nMSE :', MSE, '\nMAE :', MAE, '\nR2 :', R2, '\nadjusted R2 :', adjusted_R2) 
```
